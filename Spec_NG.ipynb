```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Analisa Gas Alam (1).xlsx',
                  engine='openpyxl', )
```

```python
df
```

```python
df.drop(['Period','H2', 'CO', 'Ar'], axis=1, inplace=True)
```

```python
df
```

```python
df_columns = df.columns.tolist()
df_columns
```

```python
for i in df_columns:
    df[i] = df[i].str.extract(r'([0-9\.]*)')
    
```

```python
df
```

```python
df = df[['n-C4H10',
 'N2',
 'H2S',
 'n-C5H12',
 'C2H6',
 'i-C4H10',
 'CO2',
 'i-C5H12',
 'CH4',
 'C6+',
 'C3H8',
         'Compressibility',
         'GHV',
          'Relative Density',
         ]]
```

```python
df
```

```python
def minMax(x):
    return pd.Series(index=['min','max'], data=[x.min(),x.max()])

df_minmax = df.apply(minMax)
```

```python
df_minmax = df_minmax.T
```

```python
with pd.ExcelWriter('NG_Spec_Pusri2B.xlsx') as writer:
    df_minmax.to_excel(writer, sheet_name='Data')
```
